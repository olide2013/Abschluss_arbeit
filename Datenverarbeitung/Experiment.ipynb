```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs  # For generating sample data

# Generate sample data (for demonstration)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Convert sample data to a DataFrame (if you have your own data, load it here)
data_df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])

# Standardize features (important for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data_df)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=0)  # Set the number of clusters to 4
kmeans.fit(X_scaled)

# Add cluster labels to the DataFrame
data_df['Cluster'] = kmeans.labels_

# Print the first few rows of the DataFrame with cluster labels
print(data_df.head())

# Plot the clusters (if data is 2D)
plt.figure(figsize=(8, 6))
plt.scatter(data_df['Feature 1'], data_df['Feature 2'], c=data_df['Cluster'], cmap='viridis', marker='o')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer
import matplotlib.pyplot as plt

# Load the DataFrame from CSV
csv_file_path = r'C:\Users\Nutzer\Desktop\BAExperiment\myproject\encoded_table.csv'
data_df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
print(data_df.head())

# Preprocess the 'CHECK DATE' column
data_df['CHECK DATE'] = pd.to_datetime(data_df['CHECK DATE'], errors='coerce')
data_df['YEAR'] = data_df['CHECK DATE'].dt.year
data_df['MONTH'] = data_df['CHECK DATE'].dt.month
data_df['DAY'] = data_df['CHECK DATE'].dt.day

# Drop the original 'CHECK DATE' column
data_df = data_df.drop(columns=['CHECK DATE'])

# Convert 'CASHED' column to numerical values (already seems to be done)
data_df['CASHED'] = data_df['CASHED'].astype(int)

# Handle categorical features
categorical_features = ['DEPARTMENT NAME', 'CONTRACT NUMBER', 'VENDOR NAME']
numerical_features = ['AMOUNT', 'YEAR', 'MONTH', 'DAY', 'CASHED']

# Preprocessing for numerical data
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])

# Apply preprocessing and clustering
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('clusterer', KMeans(n_clusters=4, random_state=0))])

# Fit the pipeline to the data
pipeline.fit(data_df)

# Add cluster labels to the original DataFrame
data_df['Cluster'] = pipeline.named_steps['clusterer'].labels_

# Print the first few rows of the DataFrame with cluster labels
print(data_df.head())

# Plot the clusters (if you have 2D or 3D features; for this example, you might use PCA for dimensionality reduction)
from sklearn.decomposition import PCA

# Reduce dimensions to 2D for plotting
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(preprocessor.transform(data_df))

plt.figure(figsize=(8, 6))
scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=data_df['Cluster'], cmap='viridis', marker='o')
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

def preprocess_data(df):
    # Convert categorical data
    categorical_features = ['DEPARTMENT NAME', 'VENDOR NAME', 'CASHED']
    le = LabelEncoder()
    
    for feature in categorical_features:
        df[feature] = le.fit_transform(df[feature].astype(str))
    
    # Convert 'CHECK DATE' to numeric
    def parse_dates(date_str):
        for fmt in ('%m/%d/%Y', '%Y'):
            try:
                return pd.to_datetime(date_str, format=fmt, errors='coerce')
            except ValueError:
                continue
        return pd.NaT

    df['CHECK DATE'] = df['CHECK DATE'].apply(parse_dates)
    df['CHECK DATE'] = df['CHECK DATE'].fillna(pd.Timestamp.now())
    df['CHECK DATE'] = df['CHECK DATE'].astype('int64') // 10**9

    # Handle missing values
    numeric_features = df.select_dtypes(include=['int64', 'float64']).columns
    categorical_features = df.select_dtypes(include=['object']).columns

    # Impute missing values for numeric features
    numeric_imputer = SimpleImputer(strategy='mean')
    df[numeric_features] = numeric_imputer.fit_transform(df[numeric_features])

    # Impute missing values for categorical features
    categorical_imputer = SimpleImputer(strategy='most_frequent')
    df[categorical_features] = categorical_imputer.fit_transform(df[categorical_features])

    # Encode categorical features
    df = pd.get_dummies(df, columns=categorical_features, drop_first=True)

    # Define label and features
    if 'label' in df.columns:
        X = df.drop(['label'], axis=1).values
        y = df['label'].values
    else:
        raise ValueError("Label column 'label' not found in the DataFrame.")
    
    return X, y
```
